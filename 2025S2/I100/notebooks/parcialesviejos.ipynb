```python
def  longest_common_substring(s1, s2):
    dp = [[0] * (1 + len(s2)) for i in range(1 + len(s1))]
    longest, x_longest = 0, 0
    for x in range(1, 1 + len(s1)):
        for y in range(1, 1 + len(s2)):
            if s1[x - 1] == s2[y - 1]:
                dp[x][y] = dp[x - 1][y - 1] + 1
                if dp[x][y] > longest:
                    longest = dp[x][y]
                    x_longest = x
    return x_longest, longest

longest_common_substring("AAABCDGH", "AEDFAAAHR")
```
