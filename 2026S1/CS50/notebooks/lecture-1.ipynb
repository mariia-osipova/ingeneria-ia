get_string function

```
string answer = get_string("What's your name? ");
```

%s is a placeholder for a string, that u don't know when u'r writing a code, but when someone else is running the code it will be filled in and substituted for their input

it's important to distinguish the comas: the firs one is just grammatically placed, but the second one tells us where is the first thing to print and where's the second one.


```
printf("hello, %s\n", answer);
```
